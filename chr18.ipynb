```python
chr18 = pd.merge(gnomAD_chr18,KGD_chr18,left_on="%POS",right_on="pos")

chr18.apply(lambda x: x.str.strip(), axis = 1)
chr18["{altreq}"] = chr18["{altreq}"].str.replace(" ", "")

chr18
```

```python
chr18_dict1 = dict(zip(chr18['%POS'], chr18['%ALT']))
chr18_dict2 = dict(zip(chr18['pos'], chr18['{altreq}']))
chr18_dict3 = dict(zip(chr18['%POS'], chr18['%REF']))

"position 같은 애들임!!!!!"
```

```python
import pandas as pd
col_names = ['position18','ref', 'g18', 'k18']
list_data = []
for key in chr18_dict1:
    if key in chr18_dict2:
        first, second = chr18_dict1[key], chr18_dict2[key]
        if first != second:
            third = second
            if key in chr18_dict3:
              a = chr18_dict3[key]
              result = [key,a,first,third]
              
            for i in ([result]):
              list_data.append(i)
print(list_data)
```

```python
df18= pd.DataFrame(list_data, columns = col_names)
print(df18)  
```
