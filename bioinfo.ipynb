```python
chr21_dict1 = dict(zip(chr21['%POS'], chr21['%ALT']))
chr21_dict2 = dict(zip(chr21['pos'], chr21['{altreq}']))
chr21_dict3 = dict(zip(chr21['%POS'], chr21['%REF']))
print(chr21_dict2)

"position 같은 애들임!!!!!"
```

```python
import pandas as pd
col_names = ['position21','ref', 'g21', 'k21']
list_data = []
for key in chr21_dict1:
    if key in chr21_dict2:
        first, second = chr21_dict1[key], chr21_dict2[key]
        if first != second:
            third = second
            if key in chr21_dict3:
              a = chr21_dict3[key]
              result = [key,a,first,third]
              
            for i in ([result]):
              list_data.append(i)
print(list_data)
```

```python
df21= pd.DataFrame(list_data, columns = col_names)
print(df21)  
```
