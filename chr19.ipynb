```python
chr19_dict1 = dict(zip(chr19['%POS'], chr19['%ALT']))
chr19_dict2 = dict(zip(chr19['pos'], chr19['{altreq}']))
chr19_dict3 = dict(zip(chr19['%POS'], chr19['%REF']))
print(chr19_dict2)

"position 같은 애들임!!!!!"
```

```python
import pandas as pd
col_names = ['position21','ref', 'g19', 'k19']
list_data = []
for key in chr19_dict1:
    if key in chr19_dict2:
        first, second = chr19_dict1[key], chr19_dict2[key]
        if first != second:
            third = second
            if key in chr19_dict3:
              a = chr19_dict3[key]
              result = [key,a,first,third]
              
            for i in ([result]):
              list_data.append(i)
print(list_data)
```

```python
df19= pd.DataFrame(list_data, columns = col_names)
print(df19)  
```
